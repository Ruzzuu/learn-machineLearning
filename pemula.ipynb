```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Example data
# data = np.array([[10,20,33,44,53,61]])
# data = np.array([[1, 2, 3, 4, 5, 6],
#                  [10, 20, 30, 40, 50, 60],
#                  [5, 15, 25, 35, 45, 55]])
data = np.array([[15],[25],[35],[45],[55],[100000]])


# Min-Max Normalization
min_max = MinMaxScaler()
data_minmax = min_max.fit_transform(data)
print("Min-Max Normalized Data:", data_minmax)

# z-score Standardization
z_scaler = StandardScaler()
data_zscore = z_scaler.fit_transform(data)
print("z-score Standardized Data:", data_zscore)

```
